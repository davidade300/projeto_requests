```python
api_base_url = 'https://api.github.com'
owner = 'amzn' # username de quem vamos extrair os dados
url = f'{api_base_url}/users/{owner}/repos'

url
```

```python
repos_list = []
for page_num in range(1, 6):
    try:
        url_page = f'{url}?page={page_num}'
        response = requests.get(url_page, headers=headers)
        repos_list.append(response.json())
    except:
        repos_list.append(None)
```

```python
repos_list
```

```python
repos_list[0][2]["name"]
```

```python
nome_repos = []
for page in repos_list: # mesma coisa que repos_list[page]
    for repo in page: # mesma coisa que repos_list[page][repo] ou seja, so falta a chave
        nome_repos.append(repo['name'])
```

```python
repos_language = []
for page in repos_list: # mesma coisa que repos_list[page]
    for repo in page: # mesma coisa que repos_list[page][repo] ou seja, so falta a chave
        repos_language.append(repo['language'])
```

```python
dados_amzn = pd.DataFrame()
dados_amzn["nome_repositorio"] = nome_repos
dados_amzn["lingua"] = repos_language
```
